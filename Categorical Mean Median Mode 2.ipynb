```python
twenty_pers_list= [var for var in df.columns if df[var].isnull().mean()*100 >20]
twenty_pers_list
```

```python
df=df.drop(columns=twenty_pers_list,axis =1)
```

```python
df.shape
```

```python
zero_perc_list= [var for var in df.columns if df[var].isnull().mean()>0]
zero_perc_list
```

```python
df_numeric= df.select_dtypes(['int64','float64'])
df_numeric.head()
```

```python
numeric_null_list= [var for var in df_numeric.columns if df_numeric[var].isnull().mean()>0]
numeric_null_list
```

```python
plt.figure(figsize=(20,10))
sb.heatmap(df_numeric.isnull())
```

```python
plt.figure(figsize=(16,9))
for i,var in enumerate(numeric_null_list):
    plt.subplot(2,2,i+1)
    sb.distplot(df_numeric[var])
```

```python
df['LotConfig'].unique()
```

```python
df_median=df.copy()

numeric_null_list=['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
categorical_list=['LotConfig','Exterior2nd','KitchenQual']
for num,cat in zip(numeric_null_list,categorical_list):
    for var_cat in df[cat].unique():
        df_median.update(df[df.loc[:,cat]==var_cat][num].replace(np.nan, df[df.loc[:,cat]==var_cat][num].median()))
```

```python
df_median.isnull().sum()
```

```python
plt.figure(figsize=(16,9))
for i,var in enumerate(numeric_null_list):
    plt.subplot(2,2,i+1)
    sb.distplot(df_numeric[var],bins=20,color='r')
    sb.distplot(df_median[var],bins=20,color='k')
```

```python
#plt.figure(figsize=(16,9))
for i,var in enumerate(numeric_null_list):
    plt.figure(figsize=(2,7))
    plt.subplot(3,1,1)
    sb.boxplot(df_numeric[var],color='r',)
    plt.grid()
    
    plt.figure(figsize=(2,7))
    plt.subplot(3,1,2)
    sb.boxplot(df_median[var],color='g')
    plt.grid()
```

```python
df_median.isnull().sum()
```

```python
df_categorical=df_median.select_dtypes(include=object)
df_categorical.head()
```

```python
zero_cat_list=[var for var in df_categorical.columns if df_categorical[var].isnull().mean()*100>0]
zero_cat_list
```

```python
df[zero_cat_list].mode()
```

```python
for var in zero_cat_list:
    df_median.update(df[var].fillna(df[var].mode()[0]))
```

```python
df_median.isnull().sum()
```

```python
plt.figure(figsize=(16,9))
for i,var in enumerate(numeric_null_list):
    plt.subplot(2,2,i+1)
    sb.distplot(df_numeric[var],bins=20,color='r')
    sb.distplot(df_median[var],bins=20,color='k')
```

```python
plt.figure(figsize=(20,4))
Mssbcls_slprice=df.groupby('MSSubClass',as_index=False)['SalePrice'].sum().sort_values(by='SalePrice',ascending=False)
barplot=sb.barplot(x='MSSubClass',y='SalePrice',data=Mssbcls_slprice)
for var in barplot.containers:
    barplot.bar_label(var)
#plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,4))
v= sb.countplot(x='SaleType',data=df_median,palette='cool')
for var in v.containers:
    v.bar_label(var)

plt.show()
```

```python
plt.figure(figsize=(10,4))
v= sb.countplot(x='SaleCondition',data=df_median,palette='autumn')
for var in v.containers:
    v.bar_label(var)

plt.show()
```
